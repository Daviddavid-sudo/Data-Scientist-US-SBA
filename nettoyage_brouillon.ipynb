incoherence car 810 entreprises ont une date de décaissement, un montant décaissé etc mais une durée de prêt = 0 et surtout que des MIS_Status radié

 GrAppv > 0 → Un montant a bien été approuvé
✅ DisbursementDate renseignée → Le prêt a été versé
✅ BalanceGross = 0 → Plus aucun montant restant
✅ MIS_Status = CHGOFF → Prêt classé en défaut
✅ Term = 0 → Aucune durée définie

INterpretation : Prêts accordés et immédiatement en défaut 

    L'argent a été débloqué mais aucun remboursement n'a eu lieu.
    Peut être des prêts annulés ou des fraudes détectées après décaissement.

Erreur dans l'encodage de Term

    Un prêt ne peut pas avoir Term = 0 (durée nulle).
    Peut-être qu’un Term court (ex: 1-3 mois) a été mal enregistré.

Prêts "fantômes" ou annulations après décaissement
